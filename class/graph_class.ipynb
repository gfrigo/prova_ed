## **GRAFOS**

**Grafos** são estruturas matemáticas utilizadas para representar **relações entre objetos**. Eles são formados por:

* **Vértices (ou nós)**: representam os objetos ou entidades.
* **Arestas (ou ligações)**: representam as conexões ou relações entre os vértices.

### Tipos de grafos

1. **Grafo não direcionado**:
   As arestas não têm direção; ou seja, se há uma ligação entre A e B, pode-se ir de A para B e de B para A.

2. **Grafo direcionado (ou dígrafo)**:
   As arestas têm direção; por exemplo, uma ligação de A para B **não** implica uma ligação de B para A.

3. **Grafo ponderado**:
   As arestas possuem **pesos** (valores numéricos), usados para representar custo, distância, tempo, etc.

4. **Grafo conexo** (apenas em grafos não direcionados):
   Existe um caminho entre **todos** os pares de vértices.

5. **Grafo cíclico e acíclico**:

   * Cíclico: contém pelo menos um ciclo (um caminho fechado).
   * Acíclico: não contém ciclos.
     Um exemplo importante é o **grafo direcionado acíclico (DAG)**, muito usado em fluxos de trabalho e dependências.

### Aplicações dos grafos

* Redes sociais (usuários como nós, amizades como arestas)
* Mapas de rotas e GPS (cidades como nós, estradas como arestas com pesos)
* Algoritmos de busca (como Dijkstra e A\*)
* Estrutura de sites (páginas como nós, links como arestas)
* Representação de moléculas em bioquímica

### Exemplo simples

Imagine 3 pessoas: Ana, Bruno e Carla.

* Ana é amiga de Bruno.
* Bruno é amigo de Carla.
* Carla não é amiga de Ana.

Podemos representar isso com um **grafo direcionado** onde os nós são as pessoas e as arestas indicam a direção da amizade.

````python
import networkx as nx
import matplotlib.pyplot as plt

# Cria um grafo não direcionado
G = nx.Graph()

# Adiciona vértices
G.add_nodes_from(["Ana", "Bruno", "Carla"])

# Adiciona arestas
G.add_edge("Ana", "Bruno")
G.add_edge("Bruno", "Carla")

# Desenha o grafo
nx.draw(G, with_labels=True, node_color='lightblue', node_size=2000, font_size=16)
plt.show()
````

````python
import networkx as nx
import matplotlib.pyplot as plt

# Cria um grafo direcionado
G = nx.DiGraph()

# Adiciona vértices
G.add_nodes_from(["Ana", "Bruno", "Carla"])

# Adiciona arestas (com direção)
G.add_edge("Ana", "Bruno")
G.add_edge("Bruno", "Carla")

# Define a posição dos nós (opcional, mas melhora a estética)
pos = nx.spring_layout(G)

# Desenha o grafo com setas
nx.draw(G, pos, with_labels=True, arrows=True,
        node_color='lightblue', node_size=2000, font_size=16,
        edge_color='gray', arrowstyle='-|>', arrowsize=20)

plt.show()
````

#**EXEMPLO:** Grafo ponderado

As arestas possuem pesos (valores numéricos), usados para representar custo, distância, tempo, etc.

````python
G = nx.Graph()
G.add_edge("A", "B", weight=4)
G.add_edge("B", "C", weight=2)
G.add_edge("C", "D", weight=10)

pos = nx.spring_layout(G)
edges = G.edges(data=True)
weights = [d['weight'] for (u, v, d) in edges]

nx.draw(G, pos, with_labels=True, node_color='lightcoral', node_size=2000, font_size=16, width=weights)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in edges})
plt.title("Grafo Ponderado")
plt.show()
````

##**EXEMPLO:** Grafo conexo

(apenas em grafos não direcionados): Existe um caminho entre todos os pares de vértices.

````python
G = nx.Graph()
G.add_edges_from([("A", "B"), ("B", "C"), ("C", "D")])

# Verifica se é conexo
print("É conexo?", nx.is_connected(G))  # True

nx.draw(G, with_labels=True, node_color='lightyellow', node_size=2000, font_size=16)
plt.title("Grafo Conexo")
plt.show()
````

##**EXEMPLO:** Grafo cíclico direcionado

Contém pelo menos um ciclo (um caminho fechado).



````python
G = nx.DiGraph()
G.add_edges_from([("A", "B"), ("B", "C"), ("C", "A")])  # ciclo: A → B → C → A

nx.draw(G, with_labels=True, node_color='lightpink', node_size=2000, font_size=16, arrows=True)
plt.title("Grafo Cíclico Direcionado")
plt.show()
````

##**EXEMPLO:** Grafo direcionado acíclico (DAG)

Não contém ciclos. Um exemplo importante é o grafo direcionado acíclico (DAG), muito usado em fluxos de trabalho e dependências.



````python
G = nx.DiGraph()
G.add_edges_from([("A", "B"), ("B", "C"), ("C", "D")])  # sem ciclos

# Verifica se é DAG
print("É acíclico?", nx.is_directed_acyclic_graph(G))  # True

nx.draw(G, with_labels=True, node_color='lightgray', node_size=2000, font_size=16, arrows=True)
plt.title("Grafo Acíclico (DAG)")
plt.show()
````

## **Por que utilizar grafos para representar relações?**

Utilizar uma **estrutura de grafos** é vantajoso porque ela permite **representar relações complexas** entre objetos de forma **clara, flexível e eficiente**. Abaixo estão as **principais vantagens**, com base em aplicações reais e fundamentos computacionais:

---

### ✅ **1. Modela relações e conexões complexas**

Grafos são ideais para representar **relacionamentos** entre entidades, o que é difícil ou ineficiente com outras estruturas (listas, vetores, matrizes).

🔍 **Exemplos práticos**:

* Rede social: pessoas (vértices), amizades (arestas)
* Rotas: cidades (vértices), estradas com distâncias (arestas com pesos)
* Internet: páginas (vértices), links (arestas)
* Biologia: proteínas, genes, neurônios, etc.

---

### ✅ **2. Suporta algoritmos eficientes**

Estruturas de grafos viabilizam algoritmos clássicos e otimizados para:

* **Busca** (BFS (Busca em Largura), DFS(Busca em Profundidade)) $\rightarrow$ PESQUISAR!
* **Caminho mínimo** (Dijkstra, A\*)
* **Agrupamento** (componentes conexos, clustering)
* **Fluxo máximo** (Ford-Fulkerson)
* **Ordenação topológica** (DAGs)

Isso torna possível resolver problemas como:

* Encontrar a menor rota entre dois pontos
* Verificar se há ciclos (dependências circulares)
* Detectar comunidades em redes sociais ou genômicas

---

### ✅ **3. Flexibilidade de representação**

Grafos podem ser:

* **Direcionados ou não**
* **Com ou sem pesos**
* **Com ou sem múltiplas arestas**
* **Dinâmicos**: vértices e arestas podem ser adicionados/removidos

Com isso, podemos representar uma enorme variedade de sistemas reais, sem necessidade de remodelar a estrutura.

---

### ✅ **4. Visualização clara**

Grafos podem ser **desenhados como diagramas**, facilitando o entendimento humano. Isso é extremamente útil no ensino, análise de redes e comunicação de modelos.

---

### ✅ **5. Aplicável em múltiplas áreas**

| Área                  | Uso de grafos                                             |
| --------------------- | --------------------------------------------------------- |
| Medicina              | Redes de interação gênica, anatomia funcional             |
| Engenharia            | Circuitos elétricos, projetos de rede                     |
| Informática           | Banco de dados orientados a grafos, redes de computadores |
| Transporte            | Otimização de rotas, logística                            |
| IA e Machine Learning | Algoritmos de recomendação, aprendizado por grafos (GNNs) |



#**EXEMPLO:**


Vamos montar um exemplo aplicado à área da saúde, mais precisamente um grafo de sintomas e doenças, simulando uma rede de diagnóstico médico. Este grafo pode ajudar estudantes a entender como sintomas se conectam a doenças e como algoritmos podem auxiliar no raciocínio clínico.

## 🧪 Exemplo: Rede de sintomas e doenças

* **Nós (vértices)**: sintomas e doenças.
* **Arestas**: ligam sintomas às doenças associadas.
* **Pesos (opcional)**: podem representar a força da associação (ex: prevalência ou probabilidade condicional).


````python
import networkx as nx
import matplotlib.pyplot as plt

# Cria um grafo bipartido direcionado: sintomas → doenças
G = nx.DiGraph()

# Sintomas
sintomas = ["Febre", "Tosse", "Dispneia", "Dor no peito", "Perda de olfato"]

# Doenças
doencas = ["COVID-19", "Gripe", "Pneumonia", "Infarto"]

# Adiciona os nós
G.add_nodes_from(sintomas, bipartite=0)
G.add_nodes_from(doencas, bipartite=1)

# Adiciona conexões entre sintomas e doenças
G.add_edge("Febre", "COVID-19")
G.add_edge("Febre", "Gripe")
G.add_edge("Febre", "Pneumonia")
G.add_edge("Tosse", "COVID-19")
G.add_edge("Tosse", "Gripe")
G.add_edge("Tosse", "Pneumonia")
G.add_edge("Dispneia", "COVID-19")
G.add_edge("Dispneia", "Pneumonia")
G.add_edge("Dor no peito", "Infarto")
G.add_edge("Dispneia", "Infarto")
G.add_edge("Perda de olfato", "COVID-19")

# Layout para separar sintomas e doenças
pos = {}
pos.update((n, (0, i)) for i, n in enumerate(sintomas))  # eixo esquerdo
pos.update((n, (2, i)) for i, n in enumerate(doencas))   # eixo direito

# Desenha o grafo
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=5000,
        font_size=8, arrows=True, edge_color='gray', arrowstyle='-|>', arrowsize=20)

plt.title("Rede de Sintomas e Doenças")
plt.axis('off')
plt.show()
````

##**ALGORITMO DE DIJKSTRA**

O **algoritmo de Dijkstra** é um algoritmo clássico de **busca de caminho mínimo** em um **grafo ponderado** (com pesos nas arestas), desenvolvido por **Edsger Dijkstra** em 1956.

---

## 🧠 **Objetivo**

Encontrar o **caminho mais curto (menor custo)** entre um **nó de origem** e todos os outros nós do grafo.

---

## ✅ **Características principais**

| Característica   | Descrição                                                                |
| ---------------- | ------------------------------------------------------------------------ |
| Tipo de grafo    | Funciona em grafos **ponderados** e **não direcionados ou direcionados** |
| Peso das arestas | Os pesos devem ser **não negativos**                                     |
| Estrutura usada  | Usa uma **fila de prioridade** (geralmente implementada com `heap`)      |
| Eficiência       | Complexidade típica: `O((V + E) log V)` com `heap`                       |

Obs:

- V = número de vértices (ou nós) no grafo

- E = número de arestas (ou ligações) no grafo
---

## 📍 **Como funciona (passo a passo)**

1. Inicializa o **custo de chegada** a todos os nós como infinito, exceto o **nó de origem** (custo = 0).
2. Marca todos os nós como **não visitados**.
3. Começa pelo nó de origem e:

   * Atualiza os **custos** dos vizinhos (se o novo caminho for mais curto).
   * Marca o nó como **visitado** (não será revisitado).
4. Escolhe o próximo nó com o **menor custo ainda não visitado**.
5. Repete até visitar todos os nós ou encontrar o destino.

---

## ✈️ **Exemplo de aplicação prática**

Imagine um hospital com os setores:

* A: Pronto-Socorro
* B: Triagem
* C: Exames
* D: Cirurgia
* E: UTI

Com distâncias (em minutos) como pesos nas conexões. Dijkstra encontra o **caminho mais rápido** do PS à UTI, considerando os tempos reais de trajeto.

---

## ✅ **Exemplo em Python com `networkx`**



### 🖨️ Saída esperada:

```
Caminho mais curto de A até E: ['A', 'B', 'C', 'D', 'E']
Custo total: 7
```

---

## 🏥 Aplicações na Saúde

| Aplicação                               | Utilidade                                                        |
| --------------------------------------- | ---------------------------------------------------------------- |
| **Rotas internas em hospitais**         | Encontrar caminho mais rápido entre setores                      |
| **Logística hospitalar**                | Otimizar entrega de medicamentos ou exames                       |
| **Análise de redes de sintomas**        | Achar a sequência de evolução mais provável de um quadro clínico |
| **Sistemas de emergência (ambulância)** | Encontrar a **rota ótima** com base em tempo/distância           |




````python
import networkx as nx

# Cria grafo com pesos
G = nx.Graph()
G.add_edge("A", "B", weight=2)
G.add_edge("A", "C", weight=5)
G.add_edge("B", "C", weight=1)
G.add_edge("B", "D", weight=4)
G.add_edge("C", "D", weight=1)
G.add_edge("D", "E", weight=3)

# Caminho mais curto de A até E
caminho = nx.dijkstra_path(G, source="A", target="E", weight="weight")
custo = nx.dijkstra_path_length(G, source="A", target="E", weight="weight")
nx.draw(G,with_labels=True, node_color='lightblue', node_size=3000,
        font_size=8, arrows=True, edge_color='gray', arrowstyle='-|>', arrowsize=20)
````

````python
print("Caminho mais curto de A até E:", caminho)
print("Custo total:", custo)
````
